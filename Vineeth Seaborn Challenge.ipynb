```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
salary_data = pd.read_csv("Salary Data.csv")
salary_data
```

```python
plt.figure()
sns.scatterplot(x='Age', y= 'Salary', data=salary_data)
plt.title('Default scatterplot Age vs salary')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='Age', y= 'Salary', hue= "Education Level", style= "Gender", data=salary_data)
plt.title('scatterplot Age vs salary by education level and gender')
plt.show()
```

```python
education_colors = {"Bachelor's" : "Green", "Masters": "Red"}
plt.figure()
sns.scatterplot(x="Age", y= "Salary", hue= "Education Level", style= "Gender", data=salary_data)
plt.title('scatterplot Age vs salary by education level and gender')
plt.show()
```

```python
for education_level in salary_data['Education Level'].unique():
    filtered_data = salary_data[salary_data['Education Level'] == education_level]
    sns.scatterplot(data=filtered_data, x='Age', y='Salary')
    plt.title(f'Scatter Plot for {education_level}')
    plt.xlim(18, 70)  # Adjust x-axis limits
    plt.ylim(20000, 120000)  # Adjust y-axis limits
    plt.show()
```

```python
sns.barplot(x="Education Level", y="Salary", data=salary_data)
plt.title("Average Salary by Education Level")
plt.show()
```

```python
sns.boxplot(x="Education Level", y="Salary", data=salary_data)
plt.title("Distribution of Salary Levels by Education Level")
plt.show()
```
